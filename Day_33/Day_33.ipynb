## Find the survival rate by Sex. Which gender had a higher chance of survival?

```python
df = pd.read_csv('titanic/train.csv')
```

```python
df.head()
```

```python
df.groupby('Sex')['Survived'].count()
```

```python
df.groupby('Sex')['Survived'].mean()
```

```python
sr = (df.groupby('Sex')['Survived'].mean())*100
```

```python
sr
```
