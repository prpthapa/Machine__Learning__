## Looping on Groups

```python
movie = pd.read_csv('imdb-top-1000.csv')
```

```python
movie.head()
```

```python
genre = movie.groupby('Genre')
```

```python
genre
```

```python
for group,data in genre:
    print(type(group),type(data))
```

```python
for group,data in genre:
    print(data)
```

```python
# find the highest rated movie from each genre

df = pd.DataFrame(columns = movie.columns)
for group,data in genre:
    top_m = data[data['IMDB_Rating'] == data['IMDB_Rating'].max()]
    df = pd.concat([df,top_m])
```

```python
df
```

## split and apply combination

```python
genre.apply('min')
```

```python
# find the number of movie starting with A for each group

def fun_ct(value):
    return value['Series_Title'].str.startswith('A').sum()
```

```python
genre.apply(fun_ct)
```

```python
# find the ranking of each movie in the group according to the IMDB score

def rank(value):
    value['Genre_rank'] = value['IMDB_Rating'].rank(ascending=False)
    return value
```

```python
genre.apply(rank)
```

```python
# find the normalized IMDB rating group wise

def norm(value):
    value['Norm_Rating'] = (value['IMDB_Rating'] - value['IMDB_Rating'].min()) / (value['IMDB_Rating'].max() - value['IMDB_Rating'].min())
    return value 
```

```python
genre.apply(norm)
```

### Group by on multiple columns

```python
# get_group

two.get_group(('Zoya Akhtar','Hrithik Roshan'))
```

```python
ckt = pd.read_csv('deliveries.csv')
```

```python
ckt
```

```python
ckt.shape
```

```python
# find the top 10 batsman in terms of runs

ckt.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(10)
```

```python
# find the batsman with max number of sixes

six = ckt[ckt['batsman_runs'] == 6]
six.groupby('batsman')['batsman_runs'].count().sort_values(ascending=False)
```

```python
ckt.head(5)
```

```python
# find the batsman with most number of 4's and 6's in last 5 over

over = ckt[ckt['over']>15]
sf = over[(over['batsman_runs'] == 6) | (over['batsman_runs'] == 4)]
sf.groupby('batsman')['batsman_runs'].count().sort_values(ascending = False).head(10)
```

```python
# find MS Dhoni runs against all the teams

temp = ckt[ckt['batsman'] == 'MS Dhoni']
temp.groupby('bowling_team')['batsman_runs'].sum().reset_index()
```

```python
def highest(batsman):
    temp = ckt[ckt['batsman'] == batsman]
    return temp.groupby('match_id')['batsman_runs'].sum().sort_values(ascending=False).head(1)
```

```python
highest('MS Dhoni')
```
