```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Colored Scatterplots

```python
iris = pd.read_csv('iris.csv')
iris.sample(6)
```

```python
iris['Species'] = iris['Species'].replace({'Iris-versicolor':0,'Iris-setosa':1,'Iris-virginica':2})
iris.sample(5)
```

```python
# cmap and (alpha -> to control opacity)

plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='jet',alpha = 0.8)
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()
plt.show()
```

# Plot Size

```python
plt.figure(figsize=(15,8))

plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='jet',alpha = 0.8)
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()
plt.show()
```

# Annotations

```python
batsman = pd.read_csv('batter.csv')
```

```python
batsman.head(4)
```

```python
df = batsman.head(100).sample(25,random_state =5 )
```

```python
df
```

```python
plt.style.use('fivethirtyeight')
```

```python
plt.scatter(df['avg'],df['strike_rate'])
plt.xlabel('Average Runs')
plt.ylabel('Strike_rate')
plt.show()
```

```python
plt.figure(figsize=(18,10))
plt.scatter(df['avg'],df['strike_rate'],s = df['runs'])
plt.xlabel('Average Runs')
plt.ylabel('Strike_rate')

for i in range(df.shape[0]):
    plt.text(df['avg'].values[i],df['strike_rate'].values[i],df['batter'].values[i])
```

# Vertical and Horizontal Lines

```python
plt.figure(figsize=(18,10))
plt.scatter(df['avg'],df['strike_rate'],s = df['runs'])
plt.xlabel('Average Runs')
plt.ylabel('Strike_rate')
plt.axhline(130,color='red')
plt.axvline(33,color='red')

for i in range(df.shape[0]):
    plt.text(df['avg'].values[i],df['strike_rate'].values[i],df['batter'].values[i])
```
