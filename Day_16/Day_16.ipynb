```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
runs = pd.read_csv('sharma-kohli.csv')
runs
```

```python
# Grid

plt.plot(runs['index'],runs['RG Sharma'],color='#0420d6')
plt.plot(runs['index'],runs['V Kohli'],color='#f52105')

plt.title('RS Sharma vs V kohli Runs Comparision')
plt.xlabel('Season')
plt.ylabel('Runs Scored')

plt.grid()
```

```python
# plt.show() -> used in the text editor to displat the plots


plt.plot(runs['index'],runs['RG Sharma'],color='#0420d6')
plt.plot(runs['index'],runs['V Kohli'],color='#f52105')

plt.title('RS Sharma vs V kohli Runs Comparision')
plt.xlabel('Season')
plt.ylabel('Runs Scored')

plt.grid()
plt.show()
```

```python
# plt.scatter on pandas data

data = pd.read_csv('batter.csv')
data = data.head(50)
data
```

```python
plt.scatter(data['avg'],data['strike_rate'],color='purple')
plt.title('Avg and SR analysis of Top 50')
plt.xlabel('Avg')
plt.ylabel('Strike_rate')
```

```python
# size 
# slower technique

tip  = sns.load_dataset('tips')
plt.scatter(tip['total_bill'],tip['tip'],s=tip['size']*20  )
```

```python
# scatterplot using the plt.plot
# faster technique
plt.plot(tip['total_bill'],tip['tip'],'o' )
```

# Bar Chart

```python
children = [20,80,60,40,50]
color = ['Red','Green','Blue','Orange','Yellow']

plt.bar(color,children,color='red')
```

```python
# horizontal bar chart

plt.barh(color,children,color='red')
```

```python
df = pd.read_csv('batsman_season_record.csv')
df
```

```python
plt.bar(np.arange(df.shape[0]) - 0.2,df['2015'],width=0.2,color='red')
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2,color='green')
plt.bar(np.arange(df.shape[0]) + 0.2,df['2017'],width=0.2,color='blue')

plt.xticks(np.arange(df.shape[0]), df['batsman'])

plt.show()
```
