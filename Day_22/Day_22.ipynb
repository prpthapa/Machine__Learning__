# Facet Plots

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
# facetplot ->only works for the figure level function only

sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter',col='smoker')
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter',row='smoker')
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter',row='smoker',col='sex')
```

# Distribution Plot
- used for univariate analysis
- used to find out the distribution
- Central Tendency
  




### Plots under distribution plots are:
- Histplot
- kdeplot
- rugplot

# Histplot

```python
# figure level -> displot
# axes level -> hisplot,kdeplot,rugplot

# using the axes level
sns.histplot(data=tips,x='total_bill')
```

```python
# using the figure level function

sns.displot(data=tips,x='total_bill',kind='hist',bins=10)
```

```python
# countplot
sns.histplot(data=tips,x='day')
```

```python
sns.displot(data=tips,x='tip',hue='sex',kind='hist')
```

```python
sns.displot(data=tips,x='tip',hue='sex',kind='hist',element='step')
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic
```

```python
sns.displot(data=titanic,x='age',hue='sex',kind='hist',element='step')
```

```python
# faceting 

sns.displot(data=titanic,x='age',col='sex',kind='hist')
```

# KDEplot

```python
titanic
```

```python
sns.displot(data=tips,x='total_bill',kind='kde')
```
