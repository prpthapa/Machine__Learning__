```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('titanic/train.csv')
```

```python
df.head()
```

## Fill missing Age values with the median age grouped by Pclass and Sex

```python
df.isnull()
```

```python
df['Age'].isnull().sum()
```

```python
df.groupby(['Pclass','Sex'])['Age'].median()
```

```python
df['Age'] = df.groupby(['Pclass', 'Sex'])['Age'].transform(
    lambda x: x.fillna(x.median())
)
```

```python
df.isnull()
```

## Extract the Fare column as a NumPy array. Compute its mean, standard deviation, and the number of passengers who paid more than the mean fare.

```python
df.head()
```

```python
fares = df['Fare'].values
fares
```

```python
means = np.mean(fares)
means
```

```python
print(np.std(fares))
```

```python
print(f"Fares who paid more the mean value is: {np.sum(means > fares)}")
```

## Using NumPy, create a new column FarePerPerson = Fare / (1 + SibSp + Parch). What is the maximum value?

```python
fare = df['Fare'].values
sib = df['SibSp'].values
par = df['Parch'].values
```

```python
df['FarePerPerson'] = fare/(1 + sib + par)
```
