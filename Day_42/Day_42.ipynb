# Bivariate & Multivariate Analysis

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

### 1. Scatterplot(numerical - numerical)

#### Bivariate analysis

```python
sns.scatterplot(x="total_bill", y="tip", data=tips)
```

#### Multivariate analysis

```python
sns.relplot(x="total_bill", y="tip", data=tips,hue="sex", style="time",size="size")
```

### 2 .Barplot(numerical - categorical)

```python
titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
sns.catplot(x="sex", y="survived", data=titanic,kind ='bar',hue="class") 
```

### 3. Boxplot(numerical - categorical)

```python
sns.catplot(x='sex', y='age', data=titanic, kind='box', hue='class')
```

### 4. Displot(numerical - categorical)

```python
sns.kdeplot(titanic[titanic['survived']==0]['age'])
sns.kdeplot(titanic[titanic['survived']==1]['age'])
```

### 5. heatmap(categorical-categorical)

```python
sns.heatmap(pd.crosstab(titanic['class'],titanic['survived']))
```

```python
titanic.groupby('class')['survived'].mean()*100
```

### 6. Clustermap(categorical - categorical)

```python
sns.clustermap(pd.crosstab(titanic['parch'],titanic['survived']))
```

### 7. Pairplot

```python
iris = sns.load_dataset("iris")
iris.head()
```

```python
sns.pairplot(iris,hue='species')
```

### 8. Lineplot

```python
flights = sns.load_dataset("flights")
flights.info()
```

```python
new = flights.groupby('year').sum(numeric_only=True).reset_index()
```

```python
sns.lineplot(x='year',y='passengers',data=new)
```
