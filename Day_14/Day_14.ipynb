## Merging


```python
# inner join

students.merge(regs,how='inner',on='student_id')
```

```python
# left join

courses.merge(regs,how='left',on='course_id')
```

```python
students.merge(regs,how='right',on='student_id')
```

```python
# outer join

students.merge(regs,how='outer',on='student_id')
```

```python
#  find the total revenue generated 

print(regs.merge(courses,how='inner',on='course_id')['price'].sum())
```

```python
# find the month by month revenue

df = pd.concat([mon1,mon2],keys=['mon1','mon2']).reset_index()
df.merge(courses,on='course_id').groupby('level_0')['price'].sum()
```

```python
# print the registration table

regs.merge(students,on='student_id').merge(courses,on='course_id')[['name','course_name','price']]
```
