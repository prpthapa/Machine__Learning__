# Concatenation

```python
b1 = np.arange(6).reshape(2,3)
b2 = np.arange(6,12).reshape(2,3)

print(b1)
print(b2)
```

```python
np.concatenate((b1,b2),axis=0)
```

```python
np.concatenate((b1,b2),axis=1)
```
