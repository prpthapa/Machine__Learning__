```python
import numpy as np
```

```python
array = np.array([[1,2,3],[4,5,6]])
print(array.shape)
```

# Datatypes as argument

```python
float32=np.array([1.32,5.78,1.33],dtype=np.float32)
float32.dtype
```

# Type Conversion


```python
boolean_array = np.array([True,False,True,False],dtype=np.bool_)
boolean_array.astype(np.int64)
```

```python
float_array = np.array([2.33,2.66,5.22,2.33],dtype=np.float32)
float_array.astype(np.int64)
```

# Indexing,iteration and slicing

```python
a = np.array([1,2,3,5])
print(a[0])
```

```python
print(a[-1])
```

```python
array_2D = np.arange(1,13).reshape(3,4)
print(array_2D)
```

#### Row index


```python
print(array_2D[0])
```

#### Column index

```python
print(array_2D[:,1])
```

### slicing


```python
print(a[2:4])
```

```python
print(array_2D[1:3])
```

```python
print(array_2D[:,2:4])
```

```python
new_array=np.arange(1,21).reshape(4,5)
print(new_array)
```

```python
print(new_array[1:3,2:4])
```

```python
for i in np.nditer(new_array):
    print(i)
```
