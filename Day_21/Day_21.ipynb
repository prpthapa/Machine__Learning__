# Seaborn
- It is simpler to use
- more graphs included

### Types of Function
- Figure level
- Axis level

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
# scatter plot -> axes level function

sns.scatterplot(data=tips,x='total_bill',y='tip',hue='sex',style='time',size='size')
```

```python
# relplot -> figure level function

sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter',hue='sex',style='time',size='size')
```

```python
import plotly as px
```

```python
# line plot


gap = px.data.gapminder()
```

```python
temp = gap[gap['country'] == 'Nepal']
```

```python
temp
```

```python
# axes level function
```

```python
sns.lineplot(data=temp,x='year',y='lifeExp')
```

```python
temp = gap[gap['country'].isin(['Nepal','Germany','Canada'])]
temp
```

```python
sns.relplot(data=temp,x='year',y='lifeExp',kind='line',hue='country',style='continent')
```
