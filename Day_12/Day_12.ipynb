```python
import numpy as np
import pandas as pd
```

```python
movie = pd.read_csv('imdb-top-1000.csv')
```

```python
movie.head()
```

# Groupby Attributes & Methods

```python
# find the total number of groups

len(movie.groupby('Genre'))
```

```python
# find items in each group

movie.groupby('Genre').size().sort_values(ascending=False)
```

```python
genre = movie.groupby('Genre')
```

```python
# first or last item from the group -> nth item

genre.first()
```

```python
genre.last()
```

```python
genre.nth(7)
```

# get_group

```python
genre.get_group('Adventure')
```

```python
# groups
genre.groups
```

```python
# describe

genre.describe()
```

```python
# sample
genre.sample()
```

```python
genre.sample(2,replace=True)
```

```python
# nunique
genre.nunique()
```

```python
# agg method -> using the dictionary

genre.agg({
    'Runtime':'mean',
    'IMDB_Rating':'mean',
    'Metascore':'max'
})
```

```python
# passing the list

genre.agg(['min','max',])
```

```python
# adding both the syntax

genre.agg({
    'Runtime':['mean','min'],
    'IMDB_Rating':'mean',
    'Metascore':['max','mean']
})
```
