```python
runs = pd.read_csv('sharma-kohli.csv')
runs
```

```python
# plotting the single plot

plt.plot(runs['index'],runs['RG Sharma'])
```

```python
# plotting multiple plots

plt.plot(runs['index'],runs['RG Sharma'])
plt.plot(runs['index'],runs['V Kohli'])
```

```python
# labels title
plt.plot(runs['index'],runs['RG Sharma'])
plt.plot(runs['index'],runs['V Kohli'])

plt.title('RS Sharma vs V kohli Runs Comparision')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# colors,line , marker(size)
plt.plot(runs['index'],runs['RG Sharma'],color='#0420d6')
plt.plot(runs['index'],runs['V Kohli'],color='#f52105')

plt.title('RS Sharma vs V kohli Runs Comparision')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
plt.plot(runs['index'],runs['RG Sharma'],color='#0420d6',linestyle='dotted')
plt.plot(runs['index'],runs['V Kohli'],color='#f52105',linestyle='dashdot')

plt.title('RS Sharma vs V kohli Runs Comparision')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
plt.plot(runs['index'],runs['RG Sharma'],color='#0420d6',linestyle='solid',linewidth=5)
plt.plot(runs['index'],runs['V Kohli'],color='#f52105',linestyle='dashed',linewidth=2)

plt.title('RS Sharma vs V kohli Runs Comparision')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# marker

# colors,line , marker(size)
plt.plot(runs['index'],runs['RG Sharma'],color='#0420d6',marker='D')
plt.plot(runs['index'],runs['V Kohli'],color='#f52105',marker='+',markersize=10)

plt.title('RS Sharma vs V kohli Runs Comparision')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# Legend
plt.plot(runs['index'],runs['RG Sharma'],color='#0420d6',label='Rohit')
plt.plot(runs['index'],runs['V Kohli'],color='#f52105',label='Virat')

plt.title('RS Sharma vs V kohli Runs Comparision')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.legend()
```
