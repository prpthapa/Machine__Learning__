# Matrix Plot
- Heatmap
- clustermap 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly as px
```

### Heatmap

```python
gap = px.data.gapminder()
```

```python
gap
```

```python
var = gap.pivot(index='country',columns='year',values='lifeExp')
```

```python
var
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(var)
```

```python
var = gap[gap['continent'] == 'Asia'].pivot(index='country',columns='year',values='lifeExp')
```

```python
var
```

```python
# annot

plt.figure(figsize=(10,10))
sns.heatmap(var,annot=True,linewidth=0.5,cmap='winter')
```
