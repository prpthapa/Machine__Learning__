# Reshaping Arrays

```python
import numpy as np
```

```python
arr1 = np.arange(1,13).reshape(3,4)
print(arr1)
```

##### convert n dimesion array into 1D array

```python
arr1.ravel()
```

```python
arr1.transpose()
```

```python
arr2=np.arange(20,32).reshape(3,4)
print(arr2)
```

```python
np.hstack((arr1,arr2))
```

```python
np.vstack((arr1,arr2))
```

```python
np.hsplit(arr2,4)
```

```python
np.vsplit(arr2,3)
```

## Fancy Indexing


```python
arr3 = np.arange(30).reshape(5,6)
arr3
```

```python
arr3[[0,2,3]]
```

```python
arr3[:,[1,2,5]]
```

## Indexing using Boolean Arrays

```python
arr4 = np.random.randint(low=1,high=80,size=16).reshape(4,4)
arr4
```

```python
arr4[(arr4>50) & (arr4%2==0)]
```
