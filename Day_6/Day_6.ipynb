# Pandas

## Panda Series

```python
import numpy as np
import pandas as pd
```

## Series from lists

```python
# string

country = ['Nepal','India','USA','Germany','China']
pd.Series(country)
```

```python
# integers
roll = [23,45,68,90,95]
pd.Series(roll)
```

```python
# custom index
marks = [42,90,58,98]
subjects = ['English','Math','Nepali','CS']

pd.Series(marks,index = subjects)
```

```python
# setting a name
pd.Series(marks,index = subjects,name='studentmarks')
```

## Series from dictionary

```python
marks ={
    'English':42,
    'Math':90,
    'Nepali':58,
    'CS':98
}
marks_series = pd.Series(marks,name='studentmarks')
marks_series
```

# Series Attributes

```python
#size
marks_series.size
```

```python
# dtype
marks_series.dtype
```

```python
# name
marks_series.name
```

```python
# is_unique

pd.Series([1,2,2,3,4,4,5]).is_unique
```

```python
# index
marks_series.index
```

```python
# values
marks_series.values
```
