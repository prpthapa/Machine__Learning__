```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# 3D Surface Plots 

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
```

```python
xx , yy = np.meshgrid(x,y)
```

```python
z = xx**2 + yy**2
```

```python
fig = plt.figure(figsize=(18,8))

ax = plt.subplot(projection='3d')

var = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(var)
```

# Contour Plots

```python
fig = plt.figure(figsize=(18,8))

ax = plt.subplot()

var = ax.contour(xx,yy,z,cmap='viridis')
fig.colorbar(var)
```

<!-- Contour fill -->

# Contour Fill

```python
fig = plt.figure(figsize=(18,8))

ax = plt.subplot()

var = ax.contourf(xx,yy,z,cmap='viridis')
fig.colorbar(var)
```

```python
ball = pd.read_csv('IPL_Ball_by_Ball_2008_2022.csv')
```

```python
ball.head()
```

```python
temp = ball[(ball['ballnumber'].isin([1,2,3,4,5,6])) & (ball['batsman_run'] == 6)]
```

```python
grid = temp.pivot_table(index='overs',columns='ballnumber',values='batsman_run',aggfunc='count')
```

```python
plt.figure(figsize=(15,5))
plt.imshow(grid)
plt.yticks(ball['overs'].unique(),list(range(1,21)))
plt.xticks(np.arange(0,6),list(range(1,7)))
plt.colorbar()
```
