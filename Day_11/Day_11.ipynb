```python
points_df = pd.DataFrame(
    {
        '1st point':[(3,4),(-6,5),(0,0),(-10,1),(4,5)],
        '2nd point':[(-3,4),(0,0),(2,2),(10,10),(1,1)]
    }
)

points_df
```

```python
def euclidean(ro):
  pt_A = ro['1st point']
  pt_B = ro['2nd point']

  return ((pt_A[0] - pt_B[0])**2 + (pt_A[1] - pt_B[1])**2)**0.5
```

```python
points_df.apply(euclidean,axis = 1 )
```

```python
points_df['distance'] = points_df.apply(euclidean,axis = 1 )
```

```python
points_df
```

# GroupBy


```python
movie = pd.read_csv('imdb-top-1000.csv')
```

```python
movie.head(5)
```

```python
genre = movie.groupby('Genre')
```

```python
# using the builtin aggregation functions

genre.sum()
```

```python
genre.min()
```

```python
genre.max()
```

```python
# top 3 genres by total earning
movie.groupby('Genre')['Gross'].sum().sort_values(ascending=False)
```

```python
movie.head()
```

```python
# find the genre with the highest Imdb rating

movie.groupby('Genre')['IMDB_Rating'].max().sort_values(ascending=False)
```

```python
# find the director with most popularity

movie.groupby('Director')['No_of_Votes'].sum().sort_values(ascending=False)
```

```python
movie.head()
```

```python
# find number of movies done by each actor

movie.groupby('Star1')['Series_Title'].count().sort_values(ascending=False)
```
