### Rows and Columns - Operations (Add, Update, Delete)

```python
data = {
    'name':['Ram','Shyam','Hari'],
    'age':[16,17,18],
    'salary':[40000,50000,60000]
           
}
```

```python
df = pd.DataFrame(data)
df
```

```python
# add new column 
df['Team'] = ['CEO','HR','CTO',]
df
```

```python
df['bonus'] = df['salary'] * 0.2
df
```

```python
# adding the rows
df.loc[len(df)] = ['ABC',19,70000,'EMP',1200]
df
```

```python
# update value - using the index
df.loc[4,'bonus'] = 3000
df
```

```python
# update value - using the name

df.loc[df.name=='Ram','salary'] = 90000
df
```

```python
# delete value - row and column values

df.drop(df[df.name == 'ABC'].index,inplace=True)
```

```python
df.drop('bonus',axis = 1)
```

```python
# sort values

df.sort_values('salary')
```
