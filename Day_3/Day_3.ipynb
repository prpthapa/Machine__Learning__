# Numpy Operations


```python
import numpy as np
```

```python
arr1= np.array([1,2,3,4,5,6])
arr2 = np.array([5,6,7,8,9,2])
```

```python
arr1-arr2
```

```python
arr1*arr2
```

```python
arr1*2
```

```python
arr1>3
```

```python
arr4 = np.arange(1,13).reshape(4,3)
arr5 = np.arange(10,22).reshape(3,4)
```

```python
arr4.dot(arr5)
```

```python
print(arr5.max())
```

```python
print(arr5.min())
```

```python
arr7 = np.arange(6).reshape(2,3)
print(arr7)
```

### Check the minimum value column-wise

```python
print(arr7.min(axis=0))
```

### Check the minimum value row-wise

```python
print(arr7.min(axis=1))
```

```python
print(arr7.sum())
```

```python
print(arr7.sum(axis=0))
```

```python
print(arr7.sum(axis=1))
```

```python
print(arr7.mean())
```

```python
print(arr7.std())
```

```python
print(np.sin(arr7))
```

```python
print(np.median(arr7))
```

```python
print(np.exp(arr7))
```
