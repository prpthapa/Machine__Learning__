```python
import seaborn as sns
```

# Categorical Estimate Plot

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
# barplot

sns.barplot(data=tips,x='day',y='total_bill')
```

```python
sns.barplot(data=tips,x='day',y='total_bill',hue='sex',errorbar=None)
```

```python
 sns.catplot(data=tips,x='day',y='total_bill',kind='bar',hue='sex',errorbar=None)
```

```python
import numpy as np
```

```python
# estimator

sns.catplot(data=tips,x='day',y='total_bill',kind='bar',hue='sex',errorbar=None,estimator=np.min)
```

```python
sns.barplot(data=tips,x='sex',y='total_bill')
```

```python
# point plot -> tells the difference between different categories

sns.pointplot(data=tips,x='sex',y='total_bill',hue='smoker',errorbar=None)
```

```python
# countplot

sns.countplot(data=tips,x='day',hue='sex')
```

```python
sns.catplot(data=tips,x='day',hue='sex',kind='count')
```

```python
# faceting using the catplot

sns.catplot(data=tips,x='day',y='total_bill',col='sex',kind='box',row='time')
```

# Regression Plot


```python
# regplot -> axes level function -> doesnot support hue parameter

sns.regplot(data=tips,x='total_bill',y='tip')
```

```python
# lmplot -> figure level function

sns.lmplot(data=tips,x='total_bill',y='tip',hue='sex')
```

```python
# residplot

sns.residplot(data=tips,x='total_bill',y='tip')
```
